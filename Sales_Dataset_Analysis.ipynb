#Data Cleaning

```python
# Handle missing values
# Impute numerical columns with median, categorical with mode
for col in ['Courier Status', 'currency', 'Amount', 'ship-city', 'ship-state', 'ship-postal-code', 'ship-country', 'promotion-ids', 'fulfilled-by']:

    col_lower = col.lower()
    if col_lower in sales_data.columns:
        if sales_data[col_lower].dtype == 'object':
            if not sales_data[col_lower].isnull().all():
                 sales_data[col_lower] = sales_data[col_lower].fillna(sales_data[col_lower].mode()[0])
        elif sales_data[col_lower].dtype in ['float64', 'int64']:
            if not sales_data[col_lower].isnull().all():
                 sales_data[col_lower] = sales_data[col_lower].fillna(sales_data[col_lower].median())


try:
  sales_data['Date'] = pd.to_datetime(sales_data['Date'])
except ValueError:

    sales_data['Date'] = sales_data['Date'].astype(str).str.replace(r'[^0-9/-]', '', regex=True)
    sales_data['Date'] = pd.to_datetime(sales_data['Date'], errors='coerce')

    sales_data.dropna(subset=['Date'], inplace=True)


if 'ship-postal-code' in sales_data.columns:
    sales_data['ship-postal-code'] = pd.to_numeric(sales_data['ship-postal-code'], errors='coerce').astype('Int64') # Use Int64 to allow NaNs


for col in ['Amount', 'Qty']:

    col_lower = col.lower()
    if col_lower in sales_data.columns:

        if pd.api.types.is_numeric_dtype(sales_data[col_lower]):
            Q1 = sales_data[col_lower].quantile(0.25)
            Q3 = sales_data[col_lower].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            # Filter out outliers
            sales_data = sales_data[(sales_data[col_lower] >= lower_bound) & (sales_data[col_lower] <= upper_bound)]
        else:
            print(f"Warning: Skipping outlier detection for non-numeric column: {col_lower}")


# Remove duplicate rows
duplicate_rows = sales_data.duplicated(keep='first')
num_duplicates = duplicate_rows.sum()
sales_data = sales_data[~duplicate_rows]

print(f"Removed {num_duplicates} duplicate rows.")

display(sales_data)
```

#Data Wrangling

```python
# Feature Engineering from 'Date' column
sales_data['Year'] = sales_data['Date'].dt.year
sales_data['Month'] = sales_data['Date'].dt.month
sales_data['Day'] = sales_data['Date'].dt.day
sales_data['DayOfWeek'] = sales_data['Date'].dt.dayofweek
sales_data['WeekOfYear'] = sales_data['Date'].dt.isocalendar().week

# Calculate Total Sales
sales_data['TotalSales'] = sales_data['Qty'] * sales_data['Amount']

sales_data['TotalSales'] = sales_data['TotalSales'].fillna(0) # fill NaN values with 0

for col in ['Year', 'Month', 'Day', 'DayOfWeek', 'WeekOfYear']:
    sales_data[col] = sales_data[col].astype(int)


categorical_cols = ['Status', 'Fulfilment', 'Sales Channel', 'ship-service-level', 'Style', 'Category', 'Size', 'currency', 'ship-city', 'ship-state', 'ship-country', 'fulfilled-by']
for col in categorical_cols:

    col_lower = col.lower()
    if col_lower in sales_data.columns:
        sales_data[col_lower] = sales_data[col_lower].astype('category')

display(sales_data.head())
```

```python
# Calculate Return Rate
returned_orders = sales_data[sales_data['Status'].str.contains('returned', na=False)]
return_rate = (len(returned_orders) / len(sales_data)) * 100

# Calculate Monthly Sales
# 'Month' and 'TotalSales' columns are now guaranteed to exist
monthly_sales = sales_data.groupby('Month')['TotalSales'].sum()

# Count Cancelled Orders
cancelled_orders_count = len(sales_data[sales_data['Status'].str.contains('cancelled', na=False)])

# Calculate Weekly Highest Sales
# 'Week' and 'TotalSales' columns are now guaranteed to exist
weekly_highest_sales = sales_data.groupby('WeekOfYear')['TotalSales'].max()

print(f"Return Rate: {return_rate:.2f}%")
print(f"Cancelled Orders: {cancelled_orders_count}")
display(monthly_sales.head())
display(weekly_highest_sales.head())
```

**Data Loading & Initial Exploration:** Loaded data, checked structure, types, and basic statistics.

**Data Cleaning**: Handled missing values using median/mode, treated outliers, and removed duplicates.

**Data Wrangling**: Created new date-related features (Year, Month, Day, etc.) and calculated 'TotalSales'.

**Key Metric Calculation:** Derived total and value of cancelled orders, total revenue, return rate, AOV, and monthly/weekly sales summaries.

**Targeted Analysis:** Analyzed sales for specific products, months (May, June), and the day of highest sales.

**Data Visualization:** Generated various plots to visualize monthly revenue, order value distribution, geographical sales trends (cities, states), monthly return rate, courier status, daily sales, shipping service levels, and sales by product size.
